```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_blobs
```

```python
Datos=make_blobs(1000,2,centers=2,cluster_std=2,random_state=2610)
```

```python
import matplotlib.pyplot as plt

plt.scatter(Datos[0][:,0],Datos[0][:,1],c=Datos[1])
plt.show()
```

```python
import pandas as pd
Dataset=pd.DataFrame(Datos[0],columns=['x','y'])
Dataset['output']=Datos[1]
Dataset
```

```python
from sklearn.metrics import classification_report
## Instancio
logistic_model=LogisticRegression()
# Ajusto
logistic_model.fit(Dataset[['x','y']],Dataset['output'])
# Predecir
y_pred=logistic_model.predict(Dataset[['x','y']])

print(classification_report(y_true=Dataset['output'],y_pred=y_pred))
```

```python
print(logistic_model.coef_)
print(logistic_model.intercept_)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_arbol=DecisionTreeClassifier(max_depth=2)

# Ajusto
model_arbol.fit(Dataset[['x','y']],Dataset['output'])
# Predecir
y_pred=model_arbol.predict(Dataset[['x','y']])

print(classification_report(y_true=Dataset['output'],y_pred=y_pred))
```
